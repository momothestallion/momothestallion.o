```python
import seaborn as sns
#sns.set()
sns.set(style="whitegrid")


import numpy as np
import pandas as pd

# importing matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nyc_df = pd.read_csv('nyc accidents.csv')
```

```python
nyc_df.head()
```

```python
nyc_df.isnull().sum()
```

```python
nyc_df.drop(['CONTRIBUTING FACTOR VEHICLE 2','CONTRIBUTING FACTOR VEHICLE 3','COLLISION_ID','VEHICLE TYPE CODE 3','VEHICLE TYPE CODE 4'],axis=1,inplace=True)
```

```python
nyc_df.rename(columns ={'CRASH DATE':'CRASH_DATE'}, inplace = True)
nyc_df.rename(columns ={'CRASH TIME':'CRASH_TIME'}, inplace = True)
nyc_df.rename(columns ={'ON STREET NAME':'ON_STREET_NAME'}, inplace = True)
nyc_df.rename(columns ={'CROSS STREET NAME':'CROSS_STREET_NAME'}, inplace = True)
nyc_df.rename(columns ={'OFF STREET NAME':'OFF_STREET_NAME'}, inplace = True)
nyc_df.rename(columns ={'CONTRIBUTING FACTOR VEHICLE 1':'CONTRIBUTING_FACTOR_VEHICLE_1'}, inplace = True)
```

```python
nyc_df.shape
```

```python
nyc_df.describe()
```

```python
#converting object to datetime (date)
nyc_df['CRASH_DATE'] = pd.to_datetime(nyc_df['CRASH_DATE'])
```

```python
#converting object to datetime (time)
nyc_df.CRASH_TIME = pd.to_datetime(nyc_df.CRASH_TIME, format='%H:%M:%S')
```

```python
# creating another column for month only(separate from date)
nyc_df['crash_month'] = pd.to_datetime(nyc_df['CRASH_DATE'],format='%M:%Y').dt.to_period('M')
```

```python
#creating another column for hours only
nyc_df['hours'] = nyc_df['CRASH_TIME'].apply(lambda x: x.hour)
```

```python
#another column for days only(separate day from date)
nyc_df['crash_day']= nyc_df.CRASH_DATE.dt.day
```

```python
nyc_df.head()
```

```python
total_accidents = nyc_df.crash_month.value_counts()
total_accidents
```

```python
plt.figure(figsize=(20,10))
plt.title('total accidents by months')
plt.pie(total_accidents,labels=total_accidents.index, autopct='%1.1f%%', startangle=180, shadow=False)
```

```python
ab = nyc_df.groupby('crash_day')['hours'].count()
ab
```

```python
nyc_df.groupby('crash_day')['hours'].value_counts().plot(kind='line', figsize=(10,5))
```

```python
street = nyc_df[['OFF_STREET_NAME','CROSS_STREET_NAME','ON_STREET_NAME']].count()
street
```

```python
plt.figure(figsize=(20,10))
plt.title('most accident street')
plt.pie(street,labels=street.index, autopct='%1.1f%%', startangle=180, shadow=False)
```
